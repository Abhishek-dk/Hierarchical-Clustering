```python
# import hierarchical clustering libraries
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sn
```

```python
Univ = pd.read_csv("Universities.csv")
```

```python
Univ
```

```python
# Normalization function 
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
# Normalized data frame (considering the numerical part of data)
#iloc use for data wnt from SAT to end
df_norm = norm_func(Univ.iloc[:,1:])
df_norm
```

```python
# create dendrogram
#single ,complex,average,centroid linkage level for creating dendrogram
dendrogram = sch.dendrogram(sch.linkage(df_norm, method='average'))
```

```python
# create clusters
hc = AgglomerativeClustering(n_clusters=3, affinity = 'euclidean',
                             linkage = 'average')
```

```python
hc
```

```python
# save clusters for chart
y_hc = hc.fit_predict(df_norm)
Clusters=pd.DataFrame(y_hc,columns=['Clusters'])
```

```python
#0 = Best college,1=med college,2=bad coll
#1=
Clusters
```

```python
df_norm['h_clusterid'] = Clusters#added new column to the data frame
```
